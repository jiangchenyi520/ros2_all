# Añadir los mundos del robot con cámara

En el directorio my_world/models añadimos el dirctorio que contiene los ficheros necesarios para simular en gazebo el modelo del robot con cámara: turtlebot3_burger_pi. Este directorio contiene los ficheros:
- model.sdf
- model.config
- y el directorio meshes con las texturas del modelo.

Podéis descargar este directorio desde [PoliformaT/Recursos/Paquetes de ROS2/turtlebot3_burger_pi.zip](https://poliformat.upv.es/access/content/group/GRA_13953_2022/Paquetes%20ROS2/turtlebot3_burger_pi.zip)

Captura_my_world_pi.PNG

Con la finalidad de poder simular diferentes entornos con el robot provisto de cámara vamos a añadir el modelo del robot con la cámara. 

Para añadirlo a nuestro entorno podemos copiar el fichero burger.model y lo renombramos burger_pi.model. En su interior cambiaremos el modelo turtlebot3_burger por turtlebot3_burger_pi.

my_world_world.PNG

````python
    <include>
      <pose>-2.0 -0.5 0.01 0.0 0.0 0.0</pose>
      <uri>model://turtlebot3_burger_pi</uri>
    </include>
````

En el fichero de lanzamiento (turtlebot3_my_world.launch.py) tendréis que cambiar el nombre del modelo:
   world_file_name = 'burger_pi.model'

# Compilar los paquetes modificados

Al terminar de hacer estos cambios tenemos que compilar el paquete modificado.

````python
colcon build --packages-select my_world
````

# Simulado el robot con cámara

Para poder simular el robot primero tendremos que exportar la variable de entorno TURTLEBOT3_MODEL con el nuevo modelo creado, y luego podemos lanzar cualquiera de los tres entornos creados ( o uno propio).

````python
#[Terminal 1]
export TURTLEBOT3_MODEL=burger_pi
ros2 launch my_world turtlebot3_my_world.launch.py
````

A continuación, podemos abrir el visualizador RVIZ para comprobar que tenemos el topic image y que la imagen se carga correctamente.

````python
#[Terminal 2]
export TURTLEBOT3_MODEL=burger_pi
rviz2 --ros-args --remap use_sim_time:=True
````

En los Displays de RVIZ añadiremos el topic image para poder ver la imagen captada desde la cámara del robot.
