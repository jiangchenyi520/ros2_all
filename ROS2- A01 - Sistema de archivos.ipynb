# Estructura de paquetes ROS 2

Es necesario crear un espacio de trabajo para trabajar con ROS. En nuestro caso vamos a trabajar en el directorio turtlebot3_ws que es donde tendremos instalados todos los paquetes de ROS 2 para trabajar con nuestro robot (TURTLEBOT3).

Si estáis trabajando en el PoliLab: recordad que debéis crear en la unidad W este espacio de trabajo.

````python
mkdir -p  $HOME/W/turtlebot3_ws/src # en Polilab
cd $HOME/W/turtlebot3_ws/src # en Polilab
````

````python
mkdir -p $HOME/turtlebot3_ws/src # en máquina virtual
cd $HOME/turtlebot3_ws/src # en máquina virtual
````

Una vez situados en nuestro espacio de trabajo podemos clonar los paquetes ROS 2 necesarios para simular nuestro robot Turtlebot3. 

````python
git clone -b galactic-devel https://github.com/ROBOTIS-GIT/turtlebot3_simulations.git
````

Tras clonar los dos paquetes debemos compilar el espacio de trabajo para que se construya la estructura de directorios adecuada para poder ejecutar las simulaciones. 

````python
source /opt/ros/galactic/setup.bash # deberemos poner esto siempre que vayamos a ejecutar un comando de ROS2 desde una terminal nueva
````

````python
cd ~/turtlebot3_ws 
colcon build --symlink-install
````

Después de la compilación deberíamos tener una estructura de directorios como la siguiente:

ROS2-estructura.drawio.png

Donde:


*   El directorio src contendrá los códigos fuente de los paquetes que iremos creando o que hemos instalado 
*   El directorio install contiene los ficheros ejecutables de los paquetes incluidos en el directorio anterior. Para poder ejecutar uno de estos paquetes deberemos ejecutar:

```
source install/setup.bash
```

*   El directorio log contendrá ficheros útiles para la depuración del código
*   El directorio build contendrá ficheros intermedios de los códigos fuentes





En el directorio src es en el que creamos los nuevos paquetes. La estructura básica de estos paquetes tendrá la siguiente forma:

_ROS2-package.drawio.png

Para crear un paquete ROS 2 deberemos situarnos en el directorio src y ejecutar la siguiente instrucción (NO HACE FALTA QUE LO HAGAIS AHORA):

````python
ros2 pkg create package_name #donde package name es el nombre del paquete a crear
````

Si estamos utilizando el Polilab tendremos que añadir los siguientes comandos cada vez que abramos un nuevo terminal:

# Incluir la configuración en el fichero de configuración de Linux

Con la finalidad de no tener que estar copiando todas las instrucciones que hacen falta para ejecutar correctamente las instrucciones de ROS vamos a introducirlas en el fichero de configuración .bashrc.
Para ello tendréis que hacer lo siguiente. 

1-Editamos el fichero de configuración .bashrc

````python
sudo nano ~/.bashrc
````

2- Añadimos las siguientes líneas al final del fichero.

````python
source /opt/ros/galactic/setup.bash # introduce en el path la ruta para ejecutar los comandos ros2
source ~/turtlebot3_ws/install/setup.bash # introduce en el path la ruta para ejecutar los paquetes que vamos a implementar y los de los ejemplos
export TURTLEBOT3_MODEL=burger # variable de entorno que contiene el modelo del robot con el que vamos a trabaajar
export ROS_DOMAIN_ID=1  # variable de entorno que contiene el dominio con el que vamos a trabajar. Aquí pondremos el número de nuestro robot
export GAZEBO_MODEL_PATH=~/turtlebot3_ws/src/turtlebot3_simulations/turtlebot3_gazebo/models:$GAZEBO_MODEL_PATH  # variable de entorno que contiene la ruta para encontrar los modelos a simular
export ROS_LOG_DIR=~/my_logs  # variable de entorno que contiene la ruta del directorio donde almancenará los mensajes de error
````
